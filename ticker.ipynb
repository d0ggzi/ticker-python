### Изменение конфигурации - разрешение видео, скорость текста, длина воспроизведения и сам текст

```python
screensize = (100,100)
message = 'There is some text that would be displayed'
speed = 80 + len(message)*13
video_duration = 3
output_name = "result"
```

### Здесь происходит непосредственно создание самого видео с заданными параметрами

```python
from moviepy.editor import *

txtClip = TextClip(message, color='white', font="Amiri-Bold",
                   kerning = 5, fontsize=80)
txtClip = txtClip.set_position(lambda t: (-speed*t, 0)).set_duration(video_duration)
cvc = CompositeVideoClip( [txtClip], size=screensize)

cvc.write_videofile(f'{output_name}.avi',fps=25,codec='mpeg4')
```
